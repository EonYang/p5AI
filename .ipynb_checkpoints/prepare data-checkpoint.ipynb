```python
from textwrap import dedent

with open('./p5_usernames.txt') as f:
    names = f.read().splitlines() 

curseWords = ['//', 'print', 'Serial', 'serial', '/*', '*/']

def hasCureseWords(elem):
    r = False
    for curseWord in curseWords:
        if curseWord in elem:
            print('find ', curseWord)
            r = True
            break
    return r
    

def cleanData(userName):
    filename = './data/' + str(userName) + '_' + 'sketch.js'
    newList = []
    print(filename)
    with open(filename) as f:
        content = f.read().splitlines()
    
    f = open(filename, 'w')
    for elem in content:
        ele = dedent(elem)
        if hasCureseWords(ele) == False and ele != '':
            newList.append(ele)     
    f.write("\n".join(newList))
    f.close()

for name in names:
    cleanData(name)
```
