```python
import json
import requests

with open('./p5_usernames.txt') as f:
    content = f.read().splitlines() 

def saveSketches(userName):
    prefix = 'https://editor.p5js.org/api/'
    suffix = '/projects'
    link = prefix + userName + suffix
    r = json.loads(requests.get(link).text)
    for i in range(len(r)):
        filename = './data/' + str(userName) + '_' + str(i) + '_' + 'sketch.js'
        print(filename)
        for file in r[i]['files']:
            if file['name'] == 'sketch.js':
                sketch = file['content']
                break
        f = open(filename, 'w+')
        f.write(sketch)
        f.close()
        

def saveInOne(userName):
    prefix = 'https://editor.p5js.org/api/'
    suffix = '/projects'
    link = prefix + userName + suffix
    filename = './data/' + str(userName) + '_' + 'sketch.js'
    print(filename)
    f = open(filename, 'w+')
    r = json.loads(requests.get(link).text)
    for i in range(len(r)):
        for file in r[i]['files']:
            if file['name'] == 'sketch.js':
                sketch = file['content']
                break
        f.write(sketch)
    f.close()
        
        
for name in content:
    saveInOne(name)

```
